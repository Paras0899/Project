```python
def Category_table():
    df = pd.read_json("testing part2 .json")
    df1 = df.transpose()
    df2 =  df1.groupby('Category').count()
    return df2

Category_table()
```

# To Print Receipt in Table form

```python
def print_Receipt_dataframe():
    with open ("testing part2 .json" , 'r') as fd:
        data = json.load(fd)
        user =[]
        Name_of_products = []
        Quantity_purchased = []
        prices_of_items_purchased = []
        while True :
            print("----type exit  in input id to break----")
            user_input = input("Enter id")
            
            if user_input == 'exit':
                break
            elif user_input in data.keys():
                
                user_name = data[user_input]["Name"]
                user_data = data[user_input]["Price per item"]
                user_data2 = input('Enter Quantity :')
                full_data = int(user_data) * int(user_data2)
                Name_of_products.append(user_name)
                Quantity_purchased.append(user_data2)
                user.append(full_data)
                prices_of_items_purchased.append(full_data)
        tz = datetime.now()
        print("===================")
        print("********************")
        print(f"Time : {tz}")
        lstt = sum(user)
        print(f'Total Bill : {lstt}')
        print("********************")
        print('====================')
        df = pd.DataFrame(list(zip(Name_of_products ,  Quantity_purchased , prices_of_items_purchased )) , columns = ["Name of Products" , "Quantity" , "Price"])
        return df
        
print_Receipt_dataframe()
```
