```python
Favorite_movie = input("Enter Movie Name :")
```

```python
title_list = df['Movie_Title'].tolist()
print(title_list)
```

# Steps For Getting Similar recommendation



```python
import difflib
```

```python
Movie_recommended = difflib.get_close_matches(Favorite_movie , title_list)
print(Movie_recommended)
```

```python
Close_match = Movie_recommended[0]
print(Close_match)
```

```python
index = df[df.Movie_Title == Close_match ]['Movie_ID'].values[0]
print(index)
```

```python
recomm_score = list(enumerate(ss[index]))
```

```python
print(recomm_score)
```

```python
len(recomm_score)
```

```python
S_S_M = sorted(recomm_score , key = lambda x:x[1] , reverse = True)
print(S_S_M)
```

# For Printing 30 Recommended movies

```python
print('Top 30 movies for you : \n ')

i = 1

for movie in S_S_M:
    new_index = movie[0]
    T_T_I = df[df.index == new_index]['Movie_Title'].values[0]
    if (i<31):
        print( i , '.' , T_T_I)
        i +=1
```

# Model For Movie Revenue Prediction

## Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
```

```python
y = df1['Movie_Revenue']
```

```python
X = df1[['Movie_Budget' , 'Movie_Popularity' , 'Movie_Runtime' , 'Movie_Vote' , 'Movie_Vote_Count']]
```

```python
X = SS.fit_transform(X)
```

# Performing Train , Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X , y , test_size = 0.3 , random_state = 2529)
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
y_predict = LR.predict(X_test)
```

# Getting Error 

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test , y_predict)
```
